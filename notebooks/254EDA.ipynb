```python
#load packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#load dataset
df_clv = pd.read_csv('filtered_data254.csv')
```

```python
#get information about the dataset
df_clv.info()
```

```python
#preview of the dataset
df_clv.head()
```

```python
#description of the dataset
df_clv.describe()
```

```python
#find the shape of the dataset
df_clv.shape
```

```python
#plot Mc and q
plt.scatter(df_clv['Mc (g/mol)'], df_clv['q (crosslink density)'])
plt.xlabel('Mc')
plt.ylabel('Crosslink Density')
plt.title('Mc vs Crosslink Density');
```

```python
#plot storage modulus and crosslink density
plt.scatter(df_clv['Storage modulus'], df_clv['q (crosslink density)'])
plt.xlabel('Storage')
plt.ylabel('Crosslink Density')
plt.title('Storage Modulus vs Crosslink Density');
```

```python
#plot histogram of crosslink density
sns.histplot(df_clv['q (crosslink density)']);
plt.xlim(0, 0.3)
```

```python
#boxplot of the formulations and the Mc
sns.boxplot(x=df_clv['Formulations'], y=df_clv['Mc (g/mol)'], data=df_clv)
plt.xlabel('Formulation')
plt.ylabel('Mc');
```

```python
sns.set_style("whitegrid");
sns.pairplot(df_clv, hue='q (crosslink density)', height=8);
plt.show()
```
