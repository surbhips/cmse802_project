```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dim = pd.read_csv('filtered_data365.csv')
```

```python
df_dim.head()
```

```python
df_dim.describe()
```

```python
#plot Mc and q
plt.scatter(df_dim['Mc (g/mol)'], df_dim['q (crosslink density)'])
plt.xlabel('Mc')
plt.ylabel('Crosslink Density')
plt.title('Mc vs Crosslink Density');
```

```python
plt.scatter(df_dim['Storage modulus (MPa)'], df_dim['q (crosslink density)'])
plt.xlabel('Storage')
plt.ylabel('Crosslink Density')
plt.title('Storage Modulus vs Crosslink Density');
```

```python
plt.scatter(df_dim['Temp (K)'], df_dim['q (crosslink density)'])
plt.xlabel('Temperature')
plt.ylabel('Crosslink Density')
plt.title('Temperature vs Crosslink Density');
```
